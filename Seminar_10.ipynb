```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df
```

```python
df.loc[df['whoAmI']=='human','good'] ='true'
df.loc[df['whoAmI']=='robot','good'] ='false'
```
